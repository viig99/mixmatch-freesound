```python
import pandas as pd
import os
import numpy as np

sub_v1 = pd.read_csv('/Users/vigi99/AirtelDEV/MixMatch-pytorch/submission/submission_v1.csv', index_col='fname')
sub_v2 = pd.read_csv('/Users/vigi99/AirtelDEV/MixMatch-pytorch/submission/submission_v2.csv', index_col='fname')
sub_v3 = pd.read_csv('/Users/vigi99/AirtelDEV/MixMatch-pytorch/submission/submission_v3.csv', index_col='fname')
test_path = os.path.abspath('/Users/vigi99/kaggle/freesound/data/test')

def labels_values(df, audio_file, name):
    predicted_labels = df[df.index == audio_file]
    top_5 = np.argsort(predicted_labels.values[0])[-5:].tolist()
    top_5.reverse()
    print("===============")
    print(name)
    print(np.array(predicted_labels.columns)[top_5])
    print(np.array(predicted_labels.values)[0][top_5])
    print("===============")
    print()
```

```python
random_audio_file = np.random.choice(sub_v1.index.values)
random_audio_file_path = os.path.join(test_path, random_audio_file)
print(random_audio_file_path)
import IPython.display as ipd
labels_values(sub_v1, random_audio_file, 'V1')
labels_values(sub_v2, random_audio_file, 'V2')
labels_values(sub_v3, random_audio_file, 'V3')
ipd.Audio(random_audio_file_path)
```
